```c++
/*
We start by including the libraries needed to run the program.
You won't need to know all the contents of these, and some libraries
are only needed to run some pre-written code you don't have to handle.
*/
#include <iostream>
#include <vector>
#include <algorithm>
#include <random>
#include <ctime>
using namespace std; //this line makes some of the code shorter

//This code is here to set up the dice box and randomize its contents
vector<char> dice_box = {'R','R','R','G','G','G','G','G','G','Y','Y','Y','Y'};
srand(time(NULL));
auto rng = default_random_engine {};
shuffle(begin(dice_box), end(dice_box), rng);

//this code is here to track any of the dice rolled to escape
vector<char> escape_dice;
```

```c++
/*
Call this function any time you need to refill the dice in the dice box
and shuffle its contents.
*/
void resetDiceBox(){
    escape_dice.clear();
    dice_box = {'R','R','R','G','G','G','G',
                             'G','G','Y','Y','Y','Y'};
    auto rng = std::default_random_engine {};
    std::shuffle(std::begin(dice_box), std::end(dice_box), rng);
}
```

```c++
//This function returns true if the box is empty, and false otherwise.
bool diceBoxIsEmpty(){
    return dice_box.size() == 0;
}
```

```c++
/*
This function returns a character representing the color of a die taken
from the dice box.
'R' = Red
'G' = Green
'Y' = Yellow
*/
char takeDie(){
    if (diceBoxIsEmpty()){
        return '0';
    }
    int elem = rand()%dice_box.size();
    char die = dice_box[elem];
    vector<char>::iterator iter1 = dice_box.begin();
    for (int i = 0; i < elem; i++){
        iter1++;
    }
    dice_box.erase(iter1);
    return die;
}
```

```c++
/*
This function rolls a die and returns the result,
represented by a character.
'H' = Hit
'E' = Escape
'B' = Brains
*/
char rollDie(char color){
    int die_roll = rand()%6+1;
    if (color == 'R'){
        if (die_roll < 4){
            return 'H';
        } else if (die_roll < 6){
            return 'E';
        } else {
            return 'B';
        }

    } else if (color == 'G'){
        if (die_roll < 4){
            return 'B';
        } else if (die_roll < 6){
            return 'E';
        } else {
            return 'H';
        }

    } else {
        if (die_roll < 3){
            return 'H';
        } else if (die_roll < 5){
            return 'E';
        } else {
            return 'B';
        }
    }
}
```

```c++
//Given a character representing the color of a die, this function prints out the word for that color
//For example, printDieColor('R') prints out "red"
void printDieColor(char color){
    if (color == 'R'){
        printf("red");
    } else if (color == 'G'){
        printf("green");
    } else if (color == 'Y'){
        printf("yellow");
    } else {
        printf("\nError: Invalid Dice Color!\n");
    }
}
```

```c++
//Given a character representing the result of a die, this function prints out the word for that result
//For example, printDieResult('B') prints out "brain"
void printDieResult(char result){
    if (result == 'B'){
        printf("brain");
    } else if (result == 'E'){
        printf("escape");
    } else if (result == 'H'){
        printf("hit");
    } else {
        printf("\nError: Invalid Dice Result!\n");
    }
}
```

```c++
void finishRound(int &brains, int &hits, int &current_player, int &escape_dice_rolled_last_throw){
    printf("Player %d is finished rolling!\n\n", current_player);
    hits = 0;
    brains = 0;
    escape_dice_rolled_last_throw = 0;
    current_player = 3 - current_player;
    resetDiceBox();
}
```

```c++
void endGame(int player1_brains, int player2_brains){
    printf("Game Over!\n");
    if (player1_brains > player2_brains){
        printf ("Player 1 wins!\n");
    } else if (player1_brains < player2_brains){
        printf ("Player 2 wins!\n");
    } else {
        printf("It's a draw!\n");
    }
}
```

```c++
void printDieInfo(int number, char result, char color){
    printf("rolling die #%d...\n",number);
    printf("You rolled a ");
    printDieResult(result);
    printf(" on a ");
    printDieColor(color);
    printf(" die.\n");
}
```

```c++
/*
This function sets aside a die that was just rolled with an 'escape' result.
It is necessary to use this in order for takeEscapeDie() to function correctly.
*/
void setAsideEscapeDie(char color){
    escape_dice.insert(escape_dice.begin(), color);
}
```

```c++
/*
This function retrieves one of the dice rolled with an "escape" result in the previous throw.
*/
char takeEscapeDie(){
    if (escape_dice.size() == 0){
        return '0';
    }
    char result = escape_dice.back();
    escape_dice.pop_back();
    return result;
}
```

```c++
/*
This function will play the Zombie Dice game.
It will rely on all the previous functions to work correctly.
*/
void playGame(){
    int current_player = 1;
    int player1_brains = 0;
    int player2_brains = 0;
    int hits = 0;
    int brains = 0;
    int escape_dice_rolled_last_throw = 0;
    resetDiceBox();

    bool is_final_turn = false;
    printf("Time for a new game!\n");
    while (true){
        int escape_dice_rolled_this_throw = 0;
        printf("It's Player %d's turn!\n", current_player);
        for (int i = 1; i < 4; i++){

            char color;
            if (escape_dice_rolled_last_throw > 0){
                printf("Re");
                escape_dice_rolled_last_throw--;
                color = takeEscapeDie();
            } else {
                color = takeDie();
            }
            if (color != '0'){
                char result = rollDie(color);

                printDieInfo(i,result,color);

                if (result == 'B'){
                    brains++;
                } else if (result == 'H'){
                    hits++;
                } else if (result == 'E'){
                    setAsideEscapeDie(color);
                    escape_dice_rolled_this_throw++;
                }
            } else {
                printf("There are no more dice left in the box!\n");
            }
        }

        escape_dice_rolled_last_throw = escape_dice_rolled_this_throw;

        if (hits >= 3){
            printf("Knocked out! No brains this round :(\n");
            finishRound(brains, hits, current_player, escape_dice_rolled_last_throw);
            if (is_final_turn){
                endGame(player1_brains, player2_brains);
                break;
            }
        }
        else {
            printf("You've acquired %d brain(s) this round. ", brains);
            printf("You've acquired %d hit(s). ", hits);
            printf("Do you want to keep rolling? Y/N");
            char answer;
            cin >> answer;
            if (answer == 'N' or answer == 'n'){

                if (current_player == 1){
                    player1_brains += brains;
                } else {
                    player2_brains += brains;
                }

                finishRound(brains, hits, current_player, escape_dice_rolled_last_throw);
                printf("Player 1 has %d brains and Player 2 has %d brains.\n",
                       player1_brains, player2_brains);

                if (is_final_turn){
                    endGame(player1_brains, player2_brains);
                    break;
                }
                else if (player1_brains > 12 || player2_brains > 12){
                    is_final_turn = true;
                    printf("Player %d has reached 13 brains!\n",3-current_player);
                    printf("Player %d gets one more turn!\n", current_player);
                }
            }
        }
    }
}
```
